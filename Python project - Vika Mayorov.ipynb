1)
כיצד השתנתה הריבית בכל אחת מהמדינות לאורך התקופה?

בחינת מגמות הריבית בישראל ובארה״ב, וזיהוי נקודות של עלייה, ירידה או יציבות במדיניות המוניטרית.

```python
Interest_Rates_df.groupby("Country")["interest_rate"].describe()
```

```python
Interest_Rates_df.groupby("Country")["interest_rate"].agg(["min", "max"])
```

```python
# הקוד יוצר גרף קווי המציג את שינויי הריבית לאורך זמן עבור כל מדינה, תוך שימוש בצבעים שונים להבחנה ביניהן,
#ומוסיף כותרת, תוויות לצירים, רשת ועיצוב לשיפור הקריאות וההצגה.
plt.figure(figsize=(12,6))

sb.lineplot(data=Interest_Rates_df, x="rate_date",y="interest_rate", hue="Country", marker="o",
            palette=["#ae2012","#0a9396"], linewidth=2.5 )

plt.title("Interest Rates Over Time",fontsize=16)
plt.xlabel("Rate Date",fontsize=12)
plt.ylabel("Interest Rate (%)",fontsize=12)
plt.grid(True)
sb.set_style("ticks")
```

```python
# הפקודה מחשבת את השינוי בריבית בין כל ערך לערך הקודם עבור כל מדינה, ומוסיפה את התוצאה כעמודה חדשה.
Interest_Rates_df["Int_rate_change"] = Interest_Rates_df.groupby("Country")["interest_rate"].diff()
```

```python
# הקוד יוצר גרף קווי המציג את השינוי בריבית לאורך זמן עבור כל מדינה, כולל קו אפס להבחנה בין עלייה לירידה,
#תוך שימוש בעיצוב לשיפור הקריאות וההשוואה.
plt.figure(figsize=(12,6))

sb.lineplot(data=Interest_Rates_df, x="rate_date",y="Int_rate_change", hue="Country", marker="o",
            palette=["#ae2012","#0a9396"], linewidth=2.5 )

plt.axhline(0, color="black", linestyle="--") 

plt.title("Change in Interest Rate Over Time",fontsize=16)
plt.xlabel("Rate Change",fontsize=12)
plt.ylabel("Date",fontsize=12)
plt.grid(True)
sb.set_style("ticks")
```

```python
# הפקודה מסננת את השורות שבהן לא חל שינוי בריבית וממיינת לפי מדינה ותאריך.
Interest_Rates_df[Interest_Rates_df["Int_rate_change"] == 0].sort_values(["Country","rate_date"])
```

```python
# הפקודה מסננת את השורות שבהן לא חל שינוי בריבית וממיינת וסוכמת לפי מדינה .
Interest_Rates_df[Interest_Rates_df["Int_rate_change"] == 0].groupby("Country").count()
```

```python
#הפקודה יוצרת גרף Boxplot המשווה את התפלגות הריבית בין המדינות, ומאפשרת לראות הבדלים ברמות הפיזור וערכים חריגים.
sb.catplot(
    data=Interest_Rates_df,
    x="Country",
    y="interest_rate",
    kind="box")
```

```python
# הפקודה יוצרת היסטוגרמה המציגה את התפלגות הריבית עבור כל מדינה, כולל קו צפיפות (KDE) להמחשת הפיזור.
sb.histplot(
    data=Interest_Rates_df,
    x="interest_rate",
    hue="Country",
    bins=10,
    kde=True)
```

```python
#הפקודות יוצרות עמודה חדשה של חודש־ושנה מתוך עמודות התאריך, כדי לאחד את הנתונים לרמת חודש ולאפשר השוואה בין הטבלאות.
Interest_Rates_df["month&year"] = Interest_Rates_df["rate_date"].dt.to_period("M")
Inflation_Data_df["month&year"] = Inflation_Data_df["inflation_date"].dt.to_period("M")
```

```python
# הפקודה מאחדת את טבלאות הריבית והאינפלציה לפי מדינה וחודש.
mergedInfInt = pd.merge(Interest_Rates_df, Inflation_Data_df, on=["Country", "month&year"], how="outer")
```

```python
#הקוד יוצר גרף פיזור עם קו רגרסיה המציג את הקשר בין ריבית לאינפלציה עבור כל מדינה, ומאפשר לזהות מגמות או קשר בין המשתנים.
sb.lmplot(
    data=mergedInfInt,
    x="interest_rate",
    y="inflation_rate",
    hue="Country",
    height=6,
    aspect=1.5
)

plt.title("Relationship Between Interest Rate and Inflation", fontsize=14)
plt.xlabel("Interest Rate (%)")
plt.ylabel("Inflation Rate (%)")

plt.show()
```

```python
#הקוד מחשב את המתאם בין ריבית לאינפלציה עבור כל מדינה, ומציג את ערך הקורלציה בצורה מסודרת לפי מדינה.

corr_total=mergedInfInt.groupby("Country")[["interest_rate", "inflation_rate"]].corr().reset_index()

corr_final=corr_total[(corr_total["level_1"] == "interest_rate")][["Country", "inflation_rate"]]

print(corr_final)
```

```python
#הקוד מחשב את המתאם בין ריבית לאינפלציה ומציג אותו כ־heatmap, כך שניתן לראות את עוצמת וכיוון הקשר בצורה ויזואלית.
corr = mergedInfInt[["interest_rate","inflation_rate"]].corr()

sb.heatmap(corr, annot=True, cmap="coolwarm")
```

4)
כיצד אירועים כלכליים וגיאופוליטיים השפיעו על האינפלציה והריבית?

ניתוח השפעת אירועים משמעותיים (כגון שינויים גלובליים או אירועים מקומיים) על המדדים, באמצעות השוואת נתונים לפני ואחרי האירוע.

```python
#הפונקציה מחשבת את השינוי בריבית ובאינפלציה סביב אירוע מסוים עבור מדינה נתונה, על ידי השוואת הערכים לפני האירוע ואחריו.
def event_impact(event_date, country):
    
    temp=mergedInfInt[mergedInfInt["Country"] == country].sort_values("month&year")
    event_month=pd.Period(event_date, freq="M")
    
    before=temp[temp["month&year"] < event_month].tail(1)
    after=temp[temp["month&year"] >= event_month].head(1)
    
    if before.empty or after.empty:
        return None
    
    return {"interest_change": after["interest_rate"].values[0] - before["interest_rate"].values[0],
            "inflation_change": after["inflation_rate"].values[0] - before["inflation_rate"].values[0]}
        
```

```python
# הקוד עובר על כל האירועים הכלכליים, מחשב עבור כל מדינה את השינוי בריבית ובאינפלציה סביב כל אירוע (כולל טיפול באירועים גלובליים),
#ושומר את התוצאות ברשימה לצורך ניתוח והצגה.
results = []

for _, row in Economic_Events_df.iterrows():  
    # אירוע גלובלי
    if pd.isna(row["country_id"]):
        countries = ["Israel", "USA"]
    
    # אירוע מקומי
    else:
        countries=["Israel"] if row["country_id"] == 1 else ["USA"]
    for country in countries:
        impact=event_impact(row["event_date"], country)
        if impact:
            results.append({
                "event": row["event_description"],
                "date": row["event_date"],
                "country": country,
                "interest_change": impact["interest_change"],
                "inflation_change": impact["inflation_change"]})
```

```python
# הקוד יוצר טבלה מהתוצאות, ממיין אותה לפי תאריך, יוצר תווית משולבת של תאריך ואירוע לצורך הצגה, ומגדיר סדר כרונולוגי להצגת האירועים בגרפים.
impact_df = pd.DataFrame(results)
impact_df = impact_df.sort_values("date")
impact_df["event_label"] = (impact_df["date"].dt.strftime("%Y-%m") + " | " + impact_df["event"])
event_order = impact_df["event_label"].drop_duplicates().tolist()
```

```python
#הקוד יוצר גרף עמודות המציג את השינוי באינפלציה סביב אירועים שונים לפי מדינה, תוך סידור כרונולוגי והוספת כותרות, קו אפס ורשת לשיפור הקריאות.
plt.figure(figsize=(14,6))

sb.barplot(
    data=impact_df,
    x="event_label",
    y="inflation_change",
    hue="country",
    order=event_order)

plt.xlabel("Event (Date | Description)", fontsize=12)
plt.ylabel("Change in Inflation Rate (%)", fontsize=12)

plt.xticks(rotation=45, ha="right")
plt.title("Impact of Events on Inflation")
plt.tight_layout()
plt.axhline(0, color="black", linestyle="--")
plt.grid(True)
plt.show()
```

```python
#הקוד יוצר גרף עמודות המציג את השינוי בריבית סביב אירועים שונים לפי מדינה, תוך סידור כרונולוגי והוספת כותרות ורשת לשיפור הקריאות.
plt.figure(figsize=(14,6))

sb.barplot(
    data=impact_df,
    x="event_label",
    y="interest_change",
    hue="country",
    order=event_order
)

plt.xlabel("Event (Date | Description)", fontsize=12)
plt.ylabel("Change in Interest Rate (%)", fontsize=12)

plt.grid(True)

plt.xticks(rotation=45, ha="right")
plt.title("Impact of Events on Interest Rate", fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# סינון אירועים גלובליים (שמופיעים בשתי המדינות) כדי לראות את השפעתם על האינפלציה בשתי המדינות
global_events = impact_df.groupby("event")["country"].nunique()
global_events = global_events[global_events > 1].index

impact_global = impact_df[impact_df["event"].isin(global_events)]

event_order_global = impact_global["event_label"].drop_duplicates().tolist()

plt.figure(figsize=(14,6))

sb.barplot(
    data=impact_global,
    x="event_label",
    y="inflation_change",
    hue="country",
    order=event_order_global)

plt.xlabel("Event (Date | Description)", fontsize=12)
plt.ylabel("Change in Inflation Rate (%)", fontsize=12)

plt.xticks(rotation=45, ha="right")
plt.title("Impact of Global Events on Inflation")
plt.axhline(0, color="black", linestyle="--")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#  סינון אירועים גלובליים (שמופיעים בשתי המדינות) כדי לראות את השפעתם על הריבית בשתי המדינות
global_events = impact_df.groupby("event")["country"].nunique()
global_events = global_events[global_events > 1].index

impact_global = impact_df[impact_df["event"].isin(global_events)]

event_order_global = impact_global["event_label"].drop_duplicates().tolist()

plt.figure(figsize=(14,6))

sb.barplot(
    data=impact_global,
    x="event_label",
    y="interest_change",
    hue="country",
    order=event_order_global)

plt.xlabel("Event (Date | Description)", fontsize=12)
plt.ylabel("Change in Inflation Rate (%)", fontsize=12)

plt.xticks(rotation=45, ha="right")
plt.title("Impact of Global Events on Inflation")
plt.axhline(0, color="black", linestyle="--")
plt.grid(True)
plt.tight_layout()
plt.show()
```
